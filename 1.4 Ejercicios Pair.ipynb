```python
#1. Verificar si un numero es negativo, positivo o cero.

numero = int(input("escribe número"))

if numero < 0:
    print("El número es negativo")

elif numero ==0:
    print ("El número es cero")

else:
    print ("El número es positivo")
```

```python
#2 Verificar si un número es par o impar

numeropar = int(input("escribe número"))

if numeropar%2 == 0:
    print("El numero es par")
else:
    print("El número es impar")
```

```python
#3. Verifica si un estudiante aprobó o suspendió un exámen aprobado >=60

notas = int(input("escribe número"))

if notas >= 60:
    print("Estás aprobado")
else:
    print("Estás suspenso")
```

```python
#4. Calcula el máximo entre dos números. Dado dos números, encuentra el máximo entre ellos.

num1 = int(input("escribe número 1"))
num2 = int(input("escribe número 2"))

if num1 > num2:
    print(num1, "es mayor que ", num2)
elif num1 == num2:
    print("Los números son iguales")
else:
    print(num2, "es mayor que", num1)
```

```python
#5. verificar si una letra es vocal o consonante

letra = input("Introduce vocal o consonante")
vocales = {'a','e','i','o','u'}

if letra.lower() in vocales:
    print("Has escrito una vocal")

else:
    print("Has introducido una consonante")
```

```python
#6. Calcular el descuento de una compra según el monto total.
#>= 100 - 10%, de lo contrario nada.

total_compra = int(input("Importe compra"))

if total_compra >= 100:
    print("El descuento aplicado de", total_compra, "sería", total_compra*0.10,"€. Por lo tanto, pagas",total_compra - (total_compra*0.10) )
else:
    print("El descuento aplicado de", total_compra, "sería 0") 
```

```python
#8 Asignar una categoria a una edad.
#<18 - "Menor de edad"
#18-64 "Adulto"
# >65 "Adulto mayor"

edad = int(input("Introduce tu edad"))

if edad <18:
    print("Categoría: Menor de edad")
elif 18<= edad <=64:
    print("Categoría: Adulto")
else:
    print("Categoría: Adulto mayor")
```

```python
#9 Verificar si tres numeros forman un triangulo rectangulo
#teorema de pitagoras 
a = int(input("Introduce número 1")) # 5
b = int(input("Introduce número 2")) #3
c = int(input("Introduce número 3")) #4

if (a^2 == (b^2+c^2)):
    print("Los números",a,b,c, "forma triángulo rectángulo")
else:
    print("Los números",a,b,c, "no forman triángulo rectángulo")
```

```python
#10 Asignar una calificación a una puntuación.
#>= 90 A
#>= 80 B
#>= 70 C
#>= 60 D
# F

calificacion = int(input("Introduce tu calificación"))

if calificacion >= 90:
    print("Tu calificación es A")
elif calificacion >= 80:
    print("Tu calificación es B")
elif calificacion >= 70:
    print("Tu calificación es C")
elif calificacion >= 60:
    print("Tu calificación es D")
else:
    print("Tu calificación es F")
```

```python
#11. Calcular el IMC. pedir peso y altura.
# IMC < 18.5 "Peso Bajo"
#IMC 18.5 - 24.9 "Peso normal"
#IMC 25-29.9 "Sobrepeso"
# Obesidad

peso = float(input("Introduce tu peso en kg")) 
altura = float(input("Introduce tu altura en metros"))

imc = round(peso /(altura*altura),2)
print (imc)

if imc < 18.5:
    print("Tienes el peso bajo", peso, altura, imc)

elif 18.5 <= imc <= 24.9:
    print("Tienes el peso normal", peso, altura, imc)

elif 25 <= imc <= 29.9:
    print("Tienes sobrepeso", peso, altura, imc) 

else:
    print("Tienes obesidad", peso, altura, imc)  
```
