```python
from datetime import date
```

```python
#1 crear clase para gestinar empleados

class Empleados:
    def __init__(self,nombre,apellidos,edad,posicion,año,vacaciones,herramientas):

        self.nombre=nombre
        self.apellidos=apellidos
        self.edad=edad
        self.posicion=posicion
        self.año=año
        self.vacaciones=vacaciones
        self.herramientas=herramientas

      
    def descripcion(self): 
        print(f"los descripcion del empleado son nombre:  {self.nombre}, posicion en la empresa: {self.posicion},año de ingreso:  {self.año},dias de vacaciones disponibles: {self.vacaciones}")

    def cambiando_pposicion(self):
        if type(self.herramientas)=='list':
            for herramienta in self.herramientas: 
                if herramienta == 'Phyton':
                    print('muy bien tienes Phyton')
                elif herramienta=='Excel':
                    print(' deja de usar Excel y aprende Phyton')
                else:
                    pass


    def calcular_antiguedad(self):
        if self.año != 0:
          
            año_actual = date.today().year
          # Cálculo de años y meses 
            años_trabajados = año_actual -self.año

            print(f'{self.nombre}, lleva {años_trabajados} años en la empresa')
           
    def solicitar_aumento_vacaciones(self,dias_solicitados):
        if dias_solicitados>0:
            self.vacaciones+=dias_solicitados
            print(f'los dias solicitados han sido aprobados, el total de dias son {self.vacaciones}' )

        else:
            print('la solicitud debe de ser en nnumero positivo')
```

```python
trabajador1=Empleados('David','Garcia',45,'administrativo',2000,30,['Phyton','Sql'])
```

```python
trabajador1.calcular_antiguedad()

```
