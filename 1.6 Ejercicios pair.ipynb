```python
#9 crear diccionario con info de una persona, 
#funcion crear_persona(nombre, edad,ciudad,estatura,color_fav), devuelva diccionario con info ingresada.

#persona1 = {"nombre":"Sara","edad":44,"ciudad":"Madrid","estatura":1.60,"color_fav":"agua-marina"}

def crear_persona(nombre,edad,ciudad,estatura,color_fav):
    ##crear diccionario con estos
    persona1 = {"nombre":nombre,"edad":edad,"ciudad":ciudad,"estatura":estatura,"color_fav":color_fav}
    return persona1

    
```

```python
persona_Sa = crear_persona("Sara",44,"Madrid",1.60,"agua-marina")
print(persona_Sa)
```

```python
persona_Lu = crear_persona("Lu",37,"Jerez",1.52,"negro")
print(persona_Lu)
```

```python
#3 determina si una cadena es palindromo en funcion es_palindromo, cadena parametro, devuelva true o false

def es_palindromo (palabra):

    if palabra[::-1] == palabra:
        return True
    else:
        return False
```

```python
es_palindromo("ana")
```

```python
es_palindromo("Lourdes")
```

```python
#4 calcular promedio de una lista de números como parametro y devuelva el promedio.

def calcular_promedio (lista_num):
    promedio = sum(lista_num)/len(lista_num)
    return promedio
```

```python
numeros = [2,5,3,4,8,9]

calcular_promedio(numeros)
```

```python
#6 validar si una palabra es una anagrama de otra. es_anagrama,(palabra1,palabra2) true, false

def es_anagrama(palabra1,palabra2): # roma, amor
    if palabra1[::-1] == palabra2: # amor==  amor
        return True
    else:
        return False
```
