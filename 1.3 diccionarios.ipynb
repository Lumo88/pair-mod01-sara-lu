```python
#CREAR DICC VACIO

clase = {}
```

```python
#crear con valores predefinidos

clase = {"luis":5,"Pedro":7,"Lu":9,"Sara":9,"Maria":4}
```

```python
#acceder a una clave especifica.

posicion = input("¿De quién quieres ver la nota?")

print(f"La nota de {posicion} es ",clase[posicion])
```

```python
#4 agregar un valor a un diccionario.

clase["Israel"] = 2

print(clase)
```

```python
clase.setdefault("Juanito",7) # ESTO ES CON PARENTESIS Y CADA ELEMENTO SEPARADO POR COMA.
print(clase)
```

```python
clase.update({"Jesus":6}) ## ESTO ES CON LLAVES Y DOS PUNTOS
print(clase)
```

```python
clase.setdefault("Juanito",4) # COMO JUANITO YA ESTÁ NO HACE NADA.
print(clase)
```

```python
clase.update({"Jesus":4}) ## SI EXISTE, MODIFICA EL VALOR
print(clase)
```

```python
clase["Israel"] = 4
print(clase)
```

```python
#elimina un par clave-valor que nos especifique por input

posicion = input("¿Quién quieres eliminar?")

print(f"El usuario {posicion} ha sido borrado ",clase.pop(posicion))
```
