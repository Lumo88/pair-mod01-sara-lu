```python
#1 generar una lista de los numero pares de 1 a 20 con list comprehension

lista_pares=[  num for num in range(1,21) if num%2==0 ]
print(lista_pares)

```

```python
#2 obtenere una lista de vocales en una cadena de texto 

cadena='quiero salir a la calle '
vocales =['a','e','i','o','u']

lista_vocales=[letra for letra in cadena if letra in vocales]
print(lista_vocales)
```

```python
#3 filtrar los numeros positivos de una lista y calcular su cuadrado

lista_numeros=[-1,5,6,-8,8,-10,-15,25]

lista_positivos=[positivos**2 for positivos in lista_numeros if positivos>0 ]
print(lista_positivos)
```
